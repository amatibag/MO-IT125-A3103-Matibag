## b. Identify columns with missing values and decide how to handle them.

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
# Create df2 from df
df2 = df.copy()

# Replace null values in Platform with 'Others'
df2['Platform'] = df2['Platform'].fillna('Others')

# Replace null values in Sentiment with 'Neutral'
df2['Sentiment'] = df2['Sentiment'].fillna('Neutral')
```

## c. Ensure columns have the appropriate data types.

```python
print(df2.dtypes)
```

```python
# Convert SignupDate to datetime format
df2['InteractionDate'] = pd.to_datetime(df2['InteractionDate'], errors='coerce')
```

```python
# unable to find a way to fix the 100 rows that has different format. Decided to drop the data instead

df2 = df2.dropna()
```

## Exploratory Data Analysis (EDA)

```python
df3 = df2.copy()

# add month column
df3['Month'] = df2['InteractionDate'].dt.month_name()

# add year column
df3['Year'] = df2['InteractionDate'].dt.year
```
